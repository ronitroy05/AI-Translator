```python



import os
import gradio as gr
import ollama
from gtts import gTTS
from gtts.lang import tts_langs  


os.system("ollama pull llama3.2:latest")


languages = {
    "English": "en",
    "Hindi": "hi",
    "French": "fr",
    "Spanish": "es",
    "Japanese": "ja",
}


available_tts_langs = tts_langs()


def translate_with_llama(text, target_lang_full):
    try:
        
        target_lang = languages[target_lang_full]

        
        prompt = (
            f"You are a professional translator with expertise in multiple languages. Your job is to accurately translate "
            f"text from one language to another while maintaining the exact meaning, tone, and context. Follow these rules strictly:\n"
            f"- Translate the text accurately into {target_lang_full}.\n"
            f"- Do not change the meaning or add unnecessary words.\n"
            f"- Do NOT provide explanations, notes, or additional text.\n"
            f"- Only return the translated text in {target_lang_full}, with no extra formatting or symbols.\n\n"
            f"Input text:\n{text}\n\n"
            f"Output (Only translated text):"
        )

       
        response = ollama.chat(model="llama3.2:latest", messages=[{"role": "user", "content": prompt}])
        translated_text = response["message"]["content"].strip()

        
        if target_lang not in available_tts_langs:
            return f" **Translated Text:** {translated_text}\n Text-to-Speech is not available for {target_lang_full}.", None

        
        audio_file = "translated_audio.mp3"
        tts = gTTS(translated_text, lang=target_lang)
        tts.save(audio_file)

        return f" **Translated Text:** {translated_text}", audio_file
    
    except Exception as e:
        return f" **Error:** {str(e)}", None


iface = gr.Interface(
    fn=translate_with_llama,
    inputs=[
        gr.Textbox(label="Enter Text to Translate"),
        gr.Dropdown(choices=list(languages.keys()), label="Select Target Language")  # Full names in dropdown
    ],
    outputs=[gr.Textbox(label="Translated Text"), gr.Audio(label="Pronunciation")],
    title=" AI-Powered Translator with LLaMA 3.2 (By: Ronit Roy)",
    description="Enter text and select a language to translate using LLaMA 3.2. The translated text will also be spoken!"
)


iface.launch()
```
